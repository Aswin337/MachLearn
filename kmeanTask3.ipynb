```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as pt
import seaborn as sns
```

```python
s=pd.read_csv("/content/video_game_sales_dataset.csv")
```

```python
s.head()
```

```python
s.isnull().sum()
```

```python
z=s.iloc[:,[7,8]].values
print(z)
```

```python
wcs=[]
for i in range(1,11):
  km=KMeans(n_clusters=i,init="k-means++",random_state=42)
  km.fit(z)
  wcs.append(km.inertia_)
```

```python
sns.set()
pt.plot(range(1,11),wcs)
pt.title("The Elbow Method")
pt.xlabel("Number of Clusters")
pt.ylabel("wcs")
pt.show()
```
