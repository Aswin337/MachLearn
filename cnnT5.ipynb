```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
Img_size=224
Batch_size=32
```

```python
tr_data=ImageDataGenerator(rescale=1./255,validation_split=0.2)
tr_gen=tr_data.flow_from_directory("/content/drive/MyDrive/btcoy",
                                   target_size=(Img_size,Img_size),
                                   batch_size=Batch_size,
                                   class_mode="binary",
                                   subset="training")
```

```python
val_gen=tr_data.flow_from_directory("/content/drive/MyDrive/btcoy",
                                   target_size=(Img_size,Img_size),
                                   batch_size=Batch_size,
                                   class_mode="binary",
                                   subset="validation")
```

```python
model=keras.Sequential([
    layers.Conv2D(32,(3,3),activation="relu",input_shape=(Img_size,Img_size,3)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64,(3,3),activation="relu"),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(128,(3,3),activation="relu"),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(128,activation="relu"),
    layers.Dense(1,activation="sigmoid")
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(tr_gen,epochs=7,validation_data=val_gen,batch_size=Batch_size)
```
